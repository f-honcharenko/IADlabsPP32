```python
led1 = led.drop(columns=['Year', 'Income composition of resources', 'Schooling', 'percentage expenditure', 'under-five deaths '])
led2 = led1.corr()
plt.figure(figsize=(16,12))

cmap = sns.diverging_palette(150, 20, s=45, l=20, n=7, center="light", as_cmap=True)
_ = sns.heatmap(led2, center=0, annot=True, fmt='.3f', square=True, cmap=cmap)
```

```python
cols = ['Hepatitis B', 'Polio', 'Diphtheria', 'BMI'] 
ledc = led[["Hepatitis B", "Polio", "Diphtheria ", " BMI "]]
ledc = ledc.cov()
plt.figure(figsize=(8,8))
plt.title('Here we can see covariance and variance coefficients', size = 16)
sns.set(font_scale=1.5)
hm = sns.heatmap(ledc,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.3f',
                 annot_kws={'size': 15},
                 cmap='gist_earth',                 
                 yticklabels=cols,
                 xticklabels=cols)
plt.tight_layout()
plt.show()
```

```python
ledc1 = led.dropna(subset=['Hepatitis B'])
ledc2 = led.dropna(subset=['Polio'])
ledc3 = led.dropna(subset=['Diphtheria '])
ledc4 = led.dropna(subset=[' thinness  1-19 years'])
sns.set(style="dark")
sns.histplot(data=ledc1, x="Hepatitis B", color="forestgreen", label="Hepatitis B", kde=True)
sns.histplot(data=ledc2, x="Polio", color="black", label="Polio", kde=True)
sns.histplot(data=ledc3, x="Diphtheria ", color="yellow", label="Diphtheria", kde=True)
plt.legend() 
plt.show()
```
